```python
import matplotlib.pyplot as plt

students_names=["Sanjay","Rahul","Karan","Wasim","Ramesh","Ajay","Sartaj","Priya"]
students_marks=[35,50,20,45,25,40,25,40]

marks_perc=[]
for x in students_marks:
  res=(x/50)*100
  marks_perc.append(res)

print(marks_perc)

def line_chart_of_student_and_marks():
  plt.plot(students_names,students_marks)
  plt.title("students marks graph")
  plt.xlabel("students names")
  plt.ylabel("students marks")
  plt.show()

line_chart_of_student_and_marks()

def bar_chart():
  plt.bar(students_names,students_marks)
  plt.title("students marks graph")
  plt.xlabel("students names")
  plt.ylabel("students marks")
  plt.show()

bar_chart()
```
